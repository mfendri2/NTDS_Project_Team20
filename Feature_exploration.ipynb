# 3.4 Degree of nodes

**For the following part, we will use our sub-sample data set to study the degree of the nodes**

```python
sub_users = pd.read_csv('datasets_final.csv')
sub_users = sub_users.fillna(value = 0)
```

```python
sub_spammer = sub_users[sub_users.label == 1]
sub_nonspammer = sub_users[sub_users.label == 0]
```

**We decide to work with boxplot, since when we use histogram, the comparison of the distributions is not as clear as using the box plot **

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 8))
medianprops = dict(linestyle='-.', linewidth=2.5, color='firebrick')
sub_users[sub_users.deg_1!= 0].boxplot(by = 'label', column = ['in_deg_1','out_deg_1','deg_1'] , ax = axes,medianprops=medianprops)
plt.suptitle("")
#plt.title("Distribution of degree of nodes for relation 1")
#plt.title("Out degree of Non Spammer and Spammer for relation 1")
#From the plot, we could see that they share similar median, and quite similar variance (variance is computed from the 
# blue horizontal line segment to the black one) The points outside the black line segment (outside the box) 
#are just outlier point 
```

```python
#Degree distribution for relation 2
# points out of the blue part is outlier. the small blue is the box and the green line is the median.
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 8))
medianprops = dict(linestyle='-.', linewidth=2.5, color='firebrick')
sub_users[sub_users.deg_2!= 0].boxplot(by = 'label', column = ['in_deg_2','out_deg_2','deg_2'] , ax = axes, medianprops=medianprops)
plt.suptitle("")
#plt.title("Out degree of Non Spammer and Spammer for relation 1")
#From the plot, we could see that they share similar median, and quite similar variance (variance is computed from the 
# blue horizontal line segment to the black one) The points outside the black line segment (outside the box) 
#are just outlier point. For non spammer, outliers spread wider than for spammer.  
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 8))
sub_users[sub_users.deg_4!= 0].boxplot(by = 'label', column = ['in_deg_4','out_deg_4','deg_4'] , ax = axes, patch_artist=True)
plt.suptitle("")
#plt.title("Out degree of Non Spammer and Spammer for relation 1")
#From the plot, we could see that they share similar median, and quite similar variance (variance is computed from the 
# blue horizontal line segment to the black one) The points outside the black line segment (outside the box) 
#are just outlier point 
```

**What we could observe from the plots is that, even though the Spammers and Non - Spammers share quite similar distribution, with low median, small variance, low mean ( which represents in the tiny box plots), their ouliers are different. The outliers of Non-Spammers could reach far higher degree than the outliers of Spammers**

**Folowing, we try another type of graph to see if it could better tell the different between spammers and non-spammers.**

```python
sub_spammer_non0 = sub_spammer[sub_spammer.in_deg_2!=0]
sub_nonspammer_non0 = sub_nonspammer[sub_nonspammer.in_deg_2!=0]
plt.figure(figsize=(10,8))
bins = np.ones_like(sub_nonspammer_non0.in_deg_2) / float(sub_nonspammer_non0.shape[0])
bins2 = np.ones_like(sub_spammer_non0.in_deg_2) / float(sub_spammer_non0.shape[0])
p1 = plt.hist(sub_nonspammer_non0.in_deg_2,bins=np.arange(min(sub_nonspammer_non0.in_deg_2), max(sub_nonspammer_non0.in_deg_2) + 1, 1),weights=bins, alpha =0.8, color ='skyblue',label = 'Non Spammer');
p2 = plt.hist(sub_spammer_non0.in_deg_2,bins=np.arange(min(sub_spammer_non0.in_deg_2), max(sub_spammer_non0.in_deg_2) + 1, 1),weights=bins2, alpha =0.5, color ='coral',label = 'Spammer');

plt.title("In degree of User nodes for relation 2",fontsize=20);
plt.xlabel("Degree ",fontsize=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Percentage",fontsize=20)
plt.legend(loc='best',fontsize = 15)
plt.show()
```
